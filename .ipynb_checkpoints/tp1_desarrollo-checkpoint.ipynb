````python
# Imports
# Importo las Bibliotecas que voy a usar en el proyecto
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# DataSources
df = pd.read_csv("data/games_march2025_reduced.csv")
````

````python
# Functions
# Definición de las funciones quevoy a usar en el proyecto.

def mostrar_valores_unicos(df, columnas):
    for col in columnas:
        print(f"\n{'='*50}")
        print(f"Columna: {col}")
        print(f"{'='*50}")

        conteo = df[col].value_counts(dropna=False)

        for valor, cantidad in conteo.items():
            print(f"{repr(valor)} -> {cantidad}")
````

````python
# es que rara la distribción de los scores de usuario
sns.histplot(df['user_score'], kde=True)
plt.show
````

````python
# es raro lo que pasa con esta variable, voy a ver ls datos de otra manera:
# conteo de cada valor único en user_score
conteo_valores = registros_con_valor['user_score'].value_counts().sort_index()
print(conteo_valores)

# Ver los primeros 20 valores más comunes
print("\nTop 10 valores más frecuentes:")
print(registros_con_valor['user_score'].value_counts().head(10))
````

````python
plt.figure(figsize=(10, 4))
sns.boxplot(x=registros_con_valor['user_score'])
plt.title('Boxplot de user_score')
plt.xlabel('user_score')
plt.show()
````

````python
# Hay una gran concentración de valores CERO (0) y algunos pocos valores arriba de 30.
````

````python
# el boxplot anterior no arroja nada útil
plt.hist(registros_con_valor['user_score'], bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('user_score')
plt.ylabel('Frecuencia')
plt.title('Histograma de user_score')
plt.show()

user_score_positivos = registros_con_valor['user_score'][registros_con_valor['user_score'] > 0]
plt.hist(user_score_positivos, bins=50, edgecolor='black', alpha=0.7, color='green')
plt.xlabel('user_score')
plt.ylabel('Frecuencia')
plt.title('user_score > 0 (sin ceros)')

plt.show()
````
