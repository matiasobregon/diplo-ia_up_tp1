# Hipótesis // preguntas
* Protocolos más utilizados

````python
# Preparación de datos
# Como voy a querer trabajar sobre alguns datos de tipo fecha, voy a tener que transformar Timestamp en cosas 
# con las que pueda trabajar

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['year'] = df['Timestamp'].dt.year
df['month'] = df['Timestamp'].dt.month
df['day'] = df['Timestamp'].dt.day
df['hour'] = df['Timestamp'].dt.hour
df['day_of_week'] = df['Timestamp'].dt.day_name()
df['month_name'] = df['Timestamp'].dt.month_name()

df.sample()
````

````python
#Hipotesis: hay algún comportamiento a lo largo del día?
attacks_by_hour = df['hour'].value_counts().sort_index()
attacks_by_day = df['day_of_week'].value_counts().sort_index()
attacks_by_year = df['year'].value_counts().sort_index()

# Contar los valores y ordenar por la categoría
months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
attacks_by_month = df['month_name'].value_counts()
attacks_by_month = attacks_by_month.reindex(months_order, fill_value=0)
````

````python
attacks_by_month
````

````python
plt.figure(figsize=(10,5))
attacks_by_month.plot(kind='bar')
plt.show()
````

````python
# Histograma básico de Packet Length
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Packet Length', bins=20)
plt.title('Distribución de Longitud de Paquetes')
plt.xlabel('Longitud del Paquete (bytes)')
plt.ylabel('Frecuencia')
plt.show()
````

````python
plt.figure(figsize=(10, 6))
plt.hist(df['Source Port'], bins=30, edgecolor='black', color='green', alpha=0.7)
plt.xlabel('Puerto de Origen')
plt.ylabel('Frecuencia')
plt.title('Distribución de Puertos de Origen')
plt.grid(True, alpha=0.3)
plt.show()
````

````python
plt.figure(figsize=(10, 6))
df['Traffic Type'].value_counts().plot(kind='bar', edgecolor='black')
plt.xlabel('Tipo de Tráfico')
plt.ylabel('Frecuencia')
plt.title('Histograma de Tipos de Tráfico')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.show()# Histograma con seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Packet Length', bins=20, kde=True)
plt.title('Distribución de Longitud de Paquetes con KDE')
plt.xlabel('Longitud del Paquete (bytes)')
plt.ylabel('Frecuencia')
plt.show()
````

````python
plt.figure(figsize=(10, 6))
df['Traffic Type'].value_counts().plot(kind='bar', edgecolor='black')
plt.xlabel('Tipo de Tráfico')
plt.ylabel('Frecuencia')
plt.title('Histograma de Tipos de Tráfico')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.show()
````
