```python
plt.figure(figsize=(6,6))
conteo = df.booking_status.value_counts()
plt.pie(conteo, labels=conteo.index, autopct="%1.1f%%", startangle=90, colors=["green","red"])
plt.title("Distribucion de booking_status")
plt.show()
```

```python
df["fecha"] = pd.to_datetime(
    df["arrival_year"].astype(str) + '-' +
    df["arrival_month"].astype(str).str.zfill(2) + '-' +
    df["arrival_date"].astype(str).str.zfill(2),
    errors='coerce'
) 
```

```python
plt.figure(figsize=(10,5))
reservas = (df.set_index("fecha").resample("D").size())
plt.plot(reservas)
plt.title("Reservas en el tiempo")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Reservas")
plt.show()
```

```python
plt.figure(figsize=(10,5))
reservas = (df.set_index("fecha").groupby("booking_status").resample("M").size().unstack(0))
plt.plot(reservas.index, reservas["Canceled"], label="Canceled", color="red")
plt.plot(reservas.index, reservas["Not_Canceled"], label="Not_Canceled", color="green")
plt.title("Reservas en el tiempo")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de Reservas")
plt.show()
```

```python
plt.figure(figsize=(10,5))
conteo_meal = df.type_of_meal_plan.value_counts() # lo modifican
plt.bar(conteo_meal.index, conteo_meal.values)
plt.title("Distribucion de reservas por meal_plan")
plt.xlabel("Meal plan")
plt.ylabel("Cantidad")
plt.show()
```

```python
def tratar_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    mean_value = df[col].mean()

    df[col+ "_clean"] = df[col].where((df[col] >= lower_bound)&(df[col]<= upper_bound), mean_value)
    return df
```

```python
df = tratar_outliers_iqr(df, "log_avg_price_per_room")
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["log_avg_price_per_room_clean"], color="salmon", edgecolor="black", kde=True)
plt.title("Distribucion de avg_price_per_room")
plt.xlabel("avg_price_per_room")
plt.ylabel("Cantidad")
plt.show()
```

h0: la variable es normal
h1: la variable no es normal

```python
from scipy.stats import shapiro # Shapiro-Wilk

stat, p_value = shapiro(df["log_avg_price_per_room_clean"])
```

```python
stat
```

```python
p_value # al 95% de confianza ( alfa = 5%) -> rechazamos la H0, es decir, la distribucion NO ES NORMAL
```

```python
from scipy.stats import kstest, norm # Kolmogorov-smirnov

data = df["log_avg_price_per_room_clean"]
data_std = (data - data.mean())/data.std()

stat, p_value = kstest(data_std, "norm")
```

```python
stat
```

```python
p_value # misma decision -> rechazamos H0, por ende, NO ES NORMAL
```
